```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as py
import seaborn as sns
```

```python
fmri = sns.load_dataset("fmri")
fmri
```

```python
fmri.head()
```

```python
sns.lineplot(x="timepoint",y="signal",data= fmri,hue = "event",style = "event",markers = True)
py.show()
```

```python
pokemon = pd.read_csv(r"C:\Users\bajpa\Downloads\pandas-master\pandas-master\pokemon_data.csv")
pokemon
```

```python
pokemon.head()
```

```python
sns.barplot(x="Legendary",y="Speed",hue="Generation",data = pokemon)
py.show()
```

```python
sns.scatterplot(x= "Attack",y="Defense",data= pokemon,hue="Type 1",style= "Type 1")
py.show()
```

```python
sns.distplot(pokemon['Attack'],color = "Green",vertical = True)
py.show()
```

```python
sns.jointplot(x= "Attack",y="Defense",data= pokemon,color="olive",kind="reg")
py.show()
```

```python
sns.boxplot(x= "Attack",y="Defense",data= pokemon)
py.show()
```

```python
sns.relplot(x= "Attack",y="Defense",data= pokemon)
py.show()
```

```python
sns.catplot(x= "Attack",y="Defense",kind="violin",data= pokemon)
py.show()
```
